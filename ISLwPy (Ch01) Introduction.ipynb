## **Chapter 1: INTRODUCTION**

#### **An Overview of Statistical Learning**
**Statistical Learning** refers to tools for data understanding, which can be classified into:

* **Supervised** statistical learning involves building a statistical model for predicting or estimating an output based on one or more inputs.

* **Unsupervised** statistical learning, that has inputs but no supervising inputs; nevertheless still can learn relationships and structure from such data.
