###**Importing the CREDIT Data**


```
#Important Library for DataFrame Processing
import numpy as np
import pandas as pd
```

```
#Showing the data frame 
url="https://raw.githubusercontent.com/dsnair/ISLR/master/data/csv/Credit.csv"
credit=pd.read_csv(url) #Saving the DataFrame to the variable names as 'credit'
credit
```

```
features=credit.columns
features
```

###**Slicing DataFrame**

#####**Selecting Data by Columns**

```
income=credit[['Income']]
income
```

```
inc_gndr_age=credit[['Income','Gender','Age']]
inc_gndr_age
```

#####**Selecting Data by Rows**

```
#Using .loc which means location of the index
credit.loc[0] #Extracting index 0 of the data
```

```
#Extracting data from multiple index 
credit.loc[100:109] 
```

#####**Selecting Data by Row and Column**

```
credit[['Income','Gender','Age']].loc[100:109]
```

#####**Selecting the Data by Condition**

```
#People whose Income is Larger than Certain Value
credit[credit['Income']>150]
```

```
#Filter Data Female Only
credit[credit['Gender']=='Female']
```

```
#Extract the DataFrame by Multiple Conditions
credit[(credit['Age']>40)&(credit['Gender']=='Female')&(credit['Income']>150)][['Income','Age','Gender','Balance']]
```

#####**DataFrame Operation**

```
#Supposed that the Income value is in USD, so to convert it to KRW, while creating new column 'IncomeKRW'
credit['IncomeKRW']=credit['Income']*1100
credit
```

#####**Calculation of Statistics of the Data**

```
#Calculating the Average value of a column
income_mean=credit['Income'].mean()
print("The Average Income is: $",income_mean)
```

```
#Finding the Median value of a column
balance_med=credit['Balance'].mean()
print("The Median of Balance is: $",balance_med)
```

```
#Showing the Oldest and Youngest Person
oldest=credit['Age'].max()
youngest=credit['Age'].min()
print("The Oldest Person's Age is: ",oldest,"y.o.","and the Youngest Person's Age is: ",youngest,"y.o")
```

```
#The Standard Deviation of Limit
limit_sd=credit['Limit'].std()
print("The Standard Deviation of the Limit is: ", limit_sd)
```

```
#The Quantiles 25% & 75%
q_25_edu=credit['Education'].quantile(0.25)
q_75_edu=credit['Education'].quantile(0.75)
print("The 25% Quantile of Education is: ",q_25_edu,"and the Quantile 75% is: ",q_75_edu)
```

```
#Print out all the statistics of dataframe
credit.describe()
```
