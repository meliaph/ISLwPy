###**The Libraries**

```
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

###**The Dataset & Its Preprocessing**

```
url="https://raw.githubusercontent.com/junyanyao/ISLR_Python/master/Dataset/Smarket.csv"
smarket=pd.read_csv(url)
smarket
```

####**Dataset Correlation Matrix**

```
smarket.corr()
```

####**Encoding the Target Variable**

```
#Response --> Change the value from Up & Down to 1 & 2
y=smarket['Direction'].map({'Up':1,'Down':0})

#Predictors 
x=smarket[['Lag1','Lag2','Lag3','Lag4','Lag5','Volume']]
```

###**The Models**

####**The Model**

```
from sklearn.linear_model import LogisticRegression
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis, QuadraticDiscriminantAnalysis
from sklearn.neighbors import KNeighborsClassifier

logreg=LogisticRegression()
lda=LinearDiscriminantAnalysis()
qda=QuadraticDiscriminantAnalysis()
knn=KNeighborsClassifier(n_neighbors=3)
```

####**The Confusion Matrix**

```
logreg.fit(x,y)
```

```
from sklearn.metrics import confusion_matrix
y_pred=logreg.predict(x)
confusion_matrix(y,y_pred)
```

```
from sklearn.metrics import accuracy_score,precision_score,recall_score
print("The Accuracy Score is: ",accuracy_score(y,y_pred))
print("The Precision Score is: ",precision_score(y,y_pred))
print("The Recall Score is: ",recall_score(y,y_pred))
print("The Mean of Response Variable is: ",np.mean(y))
```

####**Train Test Split**

```
features=['Lag1','Lag2']
train=smarket[smarket['Year']<2005]
test=smarket[smarket['Year']==2005]
x_train=train[features]
y_train=train['Direction'].map({'Up':1, 'Down':0})
x_test=test[features]
y_test=test['Direction'].map({'Up':1, 'Down':0})
```

####**LOGISTIC REGRESSION**

```
logreg.fit(x_train,y_train)
```

```
y_test_pred=logreg.predict(x_test)
confusion_matrix(y_test,y_test_pred)
```

```
print("The Accuracy Score is: ",accuracy_score(y_test,y_test_pred))
print("The Precision Score is: ",precision_score(y_test,y_test_pred))
print("The Recall Score is: ",recall_score(y_test,y_test_pred))
print("The Mean of Response Variable is: ",np.mean(y_test))
```

####**LINEAR DISCRIMINANT ANALYSIS**

```
lda.fit(x_train,y_train)
```

```
y_test_pred=lda.predict(x_test)
confusion_matrix(y_test,y_test_pred)
```

```
print("The Accuracy Score is: ",accuracy_score(y_test,y_test_pred))
print("The Precision Score is: ",precision_score(y_test,y_test_pred))
print("The Recall Score is: ",recall_score(y_test,y_test_pred))
print("The Mean of Response Variable is: ",np.mean(y_test))
```

####**QUADRATIC DISCRIMINANT ANALYSIS**

```
qda.fit(x_train,y_train)
```

```
y_test_pred=qda.predict(x_test)
confusion_matrix(y_test,y_test_pred)
```

```
print("The Accuracy Score is: ",accuracy_score(y_test,y_test_pred))
print("The Precision Score is: ",precision_score(y_test,y_test_pred))
print("The Recall Score is: ",recall_score(y_test,y_test_pred))
print("The Mean of Response Variable is: ",np.mean(y_test))
```

####**K-NEARESET NEIGHBORS**

```
knn.fit(x_train,y_train)
```

```
y_test_pred=knn.predict(x_test)
confusion_matrix(y_test,y_test_pred)
```

```
print("The Accuracy Score is: ",accuracy_score(y_test,y_test_pred))
print("The Precision Score is: ",precision_score(y_test,y_test_pred))
print("The Recall Score is: ",recall_score(y_test,y_test_pred))
print("The Mean of Response Variable is: ",np.mean(y_test))
```
